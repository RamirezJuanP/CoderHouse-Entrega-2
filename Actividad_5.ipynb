```python
import numpy as np
import ipywidgets as widgets
from IPython.display import display

def calcular(nota1, nota2, nota3, nota4, nota5):
    notas = np.array([nota1, nota2, nota3, nota4, nota5])
    promedio = np.mean(notas)
    desviacion = np.std(notas)
    mayor = np.max(notas)
    menor = np.min(notas)

    resultado.value = (f"Promedio: {promedio:.2f}\n"
                       f"Desviación estándar: {desviacion:.2f}\n"
                       f"Nota mayor: {mayor}\n"
                       f"Nota menor: {menor}")

def limpiar():
    nota1.value = 0.0
    nota2.value = 0.0
    nota3.value = 0.0
    nota4.value = 0.0
    nota5.value = 0.0
    resultado.value = ""

# Crear widgets para ingresar las notas
nota1 = widgets.FloatText(description="Nota 1:", value=0.0)
nota2 = widgets.FloatText(description="Nota 2:", value=0.0)
nota3 = widgets.FloatText(description="Nota 3:", value=0.0)
nota4 = widgets.FloatText(description="Nota 4:", value=0.0)
nota5 = widgets.FloatText(description="Nota 5:", value=0.0)

# Botón para calcular
boton_calcular = widgets.Button(description="Calcular")
boton_calcular.on_click(lambda b: calcular(nota1.value, nota2.value, nota3.value, nota4.value, nota5.value))

# Botón para limpiar
boton_limpiar = widgets.Button(description="Limpiar")
boton_limpiar.on_click(lambda b: limpiar())

# Área de salida de resultados
resultado = widgets.Textarea(value="", description="Resultados:", layout=widgets.Layout(width='100%', height='100px'))

# Mostrar la interfaz
display(nota1, nota2, nota3, nota4, nota5, boton_calcular, boton_limpiar, resultado)
```

```python
import ipywidgets as widgets
from IPython.display import display
import math

# Clase base abstracta para figuras geométricas
class FiguraGeometrica:
    def calcular_volumen(self):
        pass

    def calcular_superficie(self):
        pass

# Clase para el Cilindro
class Cilindro(FiguraGeometrica):
    def __init__(self, radio, altura):
        self.radio = radio
        self.altura = altura

    def calcular_volumen(self):
        return math.pi * self.radio**2 * self.altura

    def calcular_superficie(self):
        return 2 * math.pi * self.radio * (self.radio + self.altura)

# Clase para la Esfera
class Esfera(FiguraGeometrica):
    def __init__(self, radio):
        self.radio = radio

    def calcular_volumen(self):
        return (4/3) * math.pi * self.radio**3

    def calcular_superficie(self):
        return 4 * math.pi * self.radio**2

# Clase para la Pirámide
class Piramide(FiguraGeometrica):
    def __init__(self, base, altura, apotema):
        self.base = base
        self.altura = altura
        self.apotema = apotema

    def calcular_volumen(self):
        area_base = self.base * self.base
        return (1/3) * area_base * self.altura

    def calcular_superficie(self):
        area_base = self.base * self.base
        area_lateral = 2 * self.base * self.apotema
        return area_base + area_lateral

# Clase para la interfaz gráfica
class CalculadoraGeometrica:
    def __init__(self):
        # Crear widgets para selección de figura
        self.figura_selector = widgets.Dropdown(
            options=['Cilindro', 'Esfera', 'Pirámide'],
            description='Figura:',
            style={'description_width': 'initial'}
        )

        # Widgets para entrada de datos
        self.radio = widgets.FloatText(description='Radio (cm):', style={'description_width': 'initial'})
        self.altura = widgets.FloatText(description='Altura (cm):', style={'description_width': 'initial'})
        self.base = widgets.FloatText(description='Base (cm):', style={'description_width': 'initial'})
        self.apotema = widgets.FloatText(description='Apotema (cm):', style={'description_width': 'initial'})

        # Botón de cálculo
        self.calcular_btn = widgets.Button(description='Calcular')
        self.calcular_btn.on_click(self.on_calcular_click)

        # Widgets para resultados
        self.volumen_output = widgets.HTML(value='<b>Volumen:</b> ')
        self.superficie_output = widgets.HTML(value='<b>Superficie:</b> ')

        # Observar cambios en selector de figura
        self.figura_selector.observe(self.on_figura_change, names='value')

        # Layout inicial
        self.layout = widgets.VBox([
            self.figura_selector,
            widgets.VBox([], layout=widgets.Layout(padding='10px')),  # contenedor dinámico
            self.calcular_btn,
            self.volumen_output,
            self.superficie_output
        ])

        # Mostrar interfaz inicial
        self.on_figura_change(None)

    def on_figura_change(self, change):
        # Actualizar campos según la figura seleccionada
        campos = widgets.VBox([])
        if self.figura_selector.value == 'Cilindro':
            campos.children = [self.radio, self.altura]
        elif self.figura_selector.value == 'Esfera':
            campos.children = [self.radio]
        else:  # Pirámide
            campos.children = [self.base, self.altura, self.apotema]

        self.layout.children = [
            self.figura_selector,
            campos,
            self.calcular_btn,
            self.volumen_output,
            self.superficie_output
        ]

    def on_calcular_click(self, b):
        try:
            if self.figura_selector.value == 'Cilindro':
                figura = Cilindro(self.radio.value, self.altura.value)
            elif self.figura_selector.value == 'Esfera':
                figura = Esfera(self.radio.value)
            else:  # Pirámide
                figura = Piramide(self.base.value, self.altura.value, self.apotema.value)

            volumen = figura.calcular_volumen()
            superficie = figura.calcular_superficie()

            self.volumen_output.value = f'<b>Volumen:</b> {volumen:.2f} cm³'
            self.superficie_output.value = f'<b>Superficie:</b> {superficie:.2f} cm²'
        except Exception as e:
            self.volumen_output.value = '<b>Error:</b> Verifique los valores ingresados'
            self.superficie_output.value = ''

# Crear y mostrar la calculadora
calculadora = CalculadoraGeometrica()
display(calculadora.layout)
```
